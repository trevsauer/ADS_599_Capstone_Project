```python
import requests
import pandas as pd
from pymongo import MongoClient
import os
```

```python
# Connect to MongoDB server
client = MongoClient('mongodb://localhost:27017')

# Access a specific database (create it if it doesn't exist)
db = client['capstone']

# Access a specific collection within that database (create it if it doesn't exist)
collection = db['biomed']

```

```python
# Retrieve all documents from the collection
cursor = collection.find()

# Initialize an empty list to store the documents
documents = []

# Iterate over the cursor and append each document to the list
for document in cursor:
    documents.append(document)

# Convert the list of dictionaries into a DataFrame
df = pd.DataFrame(documents)
df
```
