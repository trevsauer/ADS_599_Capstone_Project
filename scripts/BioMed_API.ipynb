```python
import requests
import pandas as pd
```

```python
# Define the API endpoint URL
api_url = "http://api.springernature.com/meta/v2/json?q=doi:10.1007/BF00627098&api_key=cb292c878bc13f005179f8c437c73a56"
# api_key = cb292c878bc13f005179f8c437c73a56 (used at end of url)

# Make a GET request to the API
response = requests.get(api_url)

# Check if the request was successful
if response.status_code == 200:
    # Parse JSON response
    json_data = response.json()
    
    # Extract records
    records = json_data.get("records", [])
    
    # Convert records to DataFrame
    df = pd.DataFrame(records)

else:
    print("Error:", response.status_code)
    
df
```

```python
print(df.columns)
# Iterate over each column in the DataFrame
for column in df.columns:
    # Print the column name
    print("Column:", column)
    
    # Print the contents of the column
    for value in df[column]:
        print(value)
    
    # Print a separator between columns
    print("=" * 50)
```
